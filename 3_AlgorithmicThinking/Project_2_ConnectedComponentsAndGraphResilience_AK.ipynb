```python
# Project 2 - Connected components and graph resilience. Function: bfs_visited(ugraph, start_node) 
def bfs_visited(ugraph, start_node):
    """This function takes the undirected graph 
    ugraph and the node start_node and returns the set consisting 
    of all nodes that are visited by a breadth-first 
    search that starts at start_node."""
    que = []
    visited = set()
    visited.add(start_node)
    que.append(start_node)
    while len(que) != 0:
        #print "que", que
        node_tmp = que.pop(0)
        #print "enqueue", node_tmp
        for nodes in ugraph[node_tmp]:
            if nodes not in visited:
                #print "checking nodes", nodes
                visited.add(nodes)
                que.append(nodes)
            #print "visited set", visited
    return visited

print bfs_visited(GRAPH10, 0)
```

```python
# Project 2 - Connected components and graph resilience. Function: cc_visited(ugraph)
def cc_visited(ugraph):
    """This function takes the undirected graph ugraph and returns a list of sets, 
    where each set consists of all the nodes (and nothing else) in a connected component, 
    and there is exactly one set in the list for each connected component in ugraph and nothing else."""
    results = []
    
    rem_nodes = ugraph.keys()
    #print "nodes", rem_nodes
    while len(rem_nodes) != 0:
        start_node = rem_nodes.pop(0)
        #print "start_node", start_node
        visited = bfs_visited(ugraph, start_node)
        #print "visited", visited
        rem_nodes = list(set(rem_nodes).difference(visited))
        #print "rem_nodes", rem_nodes
        results.append(visited)
    return results
        
print cc_visited(GRAPH3)    
```

```python
# Project 2 - Connected components and graph resilience. Function: largest_cc_size(ugraph)
def largest_cc_size(ugraph):
    """This function takes the undirected graph ugraph and returns the size (an integer) 
    of the largest connected component in ugraph """
    visited = cc_visited(ugraph)
    size = 0
    for sets in visited:
        if len(sets) > size:
            size = len(sets)
    return size

print GRAPH3
print largest_cc_size(GRAPH3)
print GRAPH3
```

```python
"""
Provided code for application portion of module 2

Helper class for implementing efficient version
of UPA algorithm
"""

import random

class UPATrial:
    """
    Simple class to encapsulate optimizated trials for the UPA algorithm
    
    Maintains a list of node numbers with multiple instance of each number.
    The number of instances of each node number are
    in the same proportion as the desired probabilities
    
    Uses random.choice() to select a node number from this list for each trial.
    """

    def __init__(self, num_nodes):
        """
        Initialize a UPATrial object corresponding to a 
        complete graph with num_nodes nodes
        
        Note the initial list of node numbers has num_nodes copies of
        each node number
        """
        self._num_nodes = num_nodes
        self._node_numbers = [node for node in range(num_nodes) for dummy_idx in range(num_nodes)]
    
    def view(self):
        print self._node_numbers

    def run_trial(self, num_nodes):
        """
        Conduct num_nodes trials using by applying random.choice()
        to the list of node numbers
        
        Updates the list of node numbers so that each node number
        appears in correct ratio
        
        Returns:
        Set of nodes
        """
        
        # compute the neighbors for the newly-created node
        new_node_neighbors = set()
        for _ in range(num_nodes):
            new_node_neighbors.add(random.choice(self._node_numbers))
        
        # update the list of node numbers so that each node number 
        # appears in the correct ratio
        self._node_numbers.append(self._num_nodes)
        for dummy_idx in range(len(new_node_neighbors)):
            self._node_numbers.append(self._num_nodes)
        self._node_numbers.extend(list(new_node_neighbors))
        
        #update the number of nodes
        self._num_nodes += 1
        return new_node_neighbors
```

```python
# Make a dictionary corresponding to a complete directed graph with the specified number of nodes.
def make_dgraph(num_nodes):
    """ Takes the number of nodes num_nodes and returns a dictionary corresponding to a complete 
    directed graph with the specified number of nodes. """
    graph = {}
    for nodes in range(num_nodes):
        #print nodes
        temp = range(num_nodes)
        temp.remove(nodes)
        graph[nodes] = set(temp)
    return graph

# empty graph
def make_dgraph_empty(num_nodes):
    """ Takes the number of nodes num_nodes and returns a dictionary corresponding to a complete 
    directed graph with the specified number of nodes. """
    graph = {}
    for nodes in range(num_nodes):
        #print nodes
        graph[nodes] = set()
    return graph
```

```python
# Make ER graphs - undirected ER graphs, adds edges randomly 
def make_ERugraph(num_nodes, prob):
    """ Takes the number of nodes num_nodes and returns a dictionary corresponding to a random 
    undirected graph with the specified number of nodes. """
    ERugraph = make_dgraph_empty(num_nodes)
    nodes = range(num_nodes)
    #print nodes
    for node in ERugraph.keys():
        #print node
        shuffle_nodes = nodes
        random.shuffle(nodes)
        num_nodes = int(len(nodes)*prob)
        #print num_nodes, shuffle_nodes
        rand_nodes = shuffle_nodes[:num_nodes]
        #print node, "rand_nodes before", rand_nodes        
        if node in rand_nodes: rand_nodes.remove(node)
        #print 'rand_nodes after', rand_nodes
        #print 'node before', ERugraph[node]
        ERugraph[node] = ERugraph[node].union(set(rand_nodes))
        #print 'node after', ERugraph[node]
        for tmp_node in rand_nodes:
            ERugraph[tmp_node].add(node) 
            
    return ERugraph    

ERugraph1 = make_ERugraph(10, .5) 
print ERugraph1
```

```python
# generate undirected UPA graphs.
def make_UPAgraph(n_nodes, m_edges):
    """ The UPA version should add an undirected edge to the UPA graph. 
    Note that since the degree of the newly added node is no longer zero, 
    its chances of being selected in subsequent trials increases."""
    graph = make_dgraph(m_edges)
    #print graph
    for n_node in range(m_edges, n_nodes):
        all_nodes = UPATrial(n_node)
        #all_nodes.view()
        graph[n_node] = all_nodes.run_trial(m_edges)
        #print "node", n_node, "set", graph[n_node]
        for keys in graph[n_node]: 
            #print "node update with key", keys, graph[keys]
            graph[keys].add(n_node)
            #print "node update done", graph[keys]
        #print graph
    return graph
    
UPAgraph1 = make_UPAgraph(10, 5)
print UPAgraph1
```

```python
def targeted_order2(ugraph):
    """
    Compute a targeted attack order consisting
    of nodes of maximal degree
    
    Returns:
    A list of nodes
    """
    # copy the graph
    new_graph = copy_graph(ugraph)
    
    order = []    
    while len(new_graph) > 0:
        max_degree = -1
        for node in new_graph:
            if len(new_graph[node]) > max_degree:
                max_degree = len(new_graph[node])
                max_degree_node = node
        
        neighbors = new_graph[max_degree_node]
        new_graph.pop(max_degree_node)
        for neighbor in neighbors:
            new_graph[neighbor].remove(max_degree_node)

        order.append(max_degree_node)
    return order
```

```python
# fast_targeted_order
def fast_targeted_order(graph):
    """
    Fast compute a targeted attack order consisting
    of nodes of maximal degree
    
    Returns:
    A list of nodes
    """
    n_graph = copy_graph(graph)
    
    n_nodes = len(n_graph.keys())
    degree_sets = [set() for nodes in range(n_nodes)]
    #print degree_sets
    for node in range(n_nodes):
        degree = len(n_graph[node])
        degree_sets[degree].add(node)
    #print degree_sets
    l_degree = []
    #i = 0
    #print "Number of nodes", n_nodes
    for node_tmp in range(n_nodes-1, -1, -1):
        while len(degree_sets[node_tmp]) != 0:
            l_nodes = list(degree_sets[node_tmp])
            #print "degree", node_tmp, "list of nodes", l_nodes
            r_node = random.choice(l_nodes)
            #print "random node", r_node
            degree_sets[node_tmp].remove(r_node)
            #print 'degree_sets[node_tmp]', degree_sets[node_tmp]
            #print 'n_graph[r_node]', n_graph[r_node]
            for neighbor in n_graph[r_node]:
                #print 'node', neighbor
                degree_tmp = len(n_graph[neighbor])
                #print 'degree_tmp', degree_tmp
                #print 'degrees -1, 0', degree_sets[degree_tmp-1], degree_sets[degree_tmp]
                degree_sets[degree_tmp-1].add(neighbor)
                degree_sets[degree_tmp].remove(neighbor)
                #print 'degrees -1, 0', degree_sets[degree_tmp-1], degree_sets[degree_tmp]
            delete_node(n_graph, r_node)
            l_degree.append(r_node)
            #print n_graph
    return l_degree

print GRAPH10
print targeted_order2(GRAPH10)
print fast_targeted_order(GRAPH10)
```
